```python
import pandas as pd
import numpy as np
```

```python
data = {'names' : ['muhammad', 'faisal', '', 'ayesha', 'f#araz'],
        'number' : ['2', '3', '4', '', '6'],
        'dob': ['1/1/1990', '2/1/1995', '5/13/1996', '1/2/2000', '31/12/2001'],
        'salary': ['20000', '4000', '4thousand', '500000', '2000']}
```

```python
data = pd.DataFrame(data = data, columns = ['names', 'number', 'dob', 'salary'])
data
```

```python
data = data.replace({'names' : {'' : 'FAST'}})
data
```

```python
data.iloc[4, 0] = data.iloc[4, 0].strip().replace("#", "")
data
```

```python
data['number'] = pd.to_numeric(data['number'])
data['number'].fillna(round(data['number'].mean(), 2), inplace = True)
data
```

```python
data = data.replace({'salary' : {'4thousand' : '4000'}})
data
```

```python
data['dob'] = pd.to_datetime(data['dob'], errors='coerce')
data
```

```python
data['salary'] = pd.to_numeric(data['salary'])

print("Mean : ", data['salary'].mean())
print("St. dev : ", data['salary'].std())
print("Median : ", data['salary'].median())
print("Max : ", data['salary'].max())
```
